```python
import pandas as pd
import psycopg2
from psycopg2 import sql
from google.colab import drive

# Configuración de la base de datos
DB_HOST = "ep-snowy-dawn-a8omlpke-pooler.eastus2.azure.neon.tech"
DB_NAME = "neondb"
DB_USER = "neondb_owner"
DB_PASSWORD = "npg_BcpnId62FYhS"

# Conectar a la base de datos
def get_db_connection():
    return psycopg2.connect(
        host=DB_HOST,
        dbname=DB_NAME,
        user=DB_USER,
        password=DB_PASSWORD,
        sslmode="require"
    )

# Montar Google Drive y leer el fichero CSV
drive.mount('/content/drive')
csv_file_path = '/content/drive/My Drive/Lotto_it.csv'

# Leer el CSV
try:
    df = pd.read_csv(csv_file_path, sep=';', encoding='ISO-8859-1')
    print(f"Archivo CSV cargado con éxito. Número de filas: {len(df)}")
except Exception as e:
    print(f"Error al leer el archivo CSV: {e}")

# Conectar a la base de datos
try:
    conn = get_db_connection()
    cursor = conn.cursor()
    print("Conexión a la base de datos establecida correctamente.")
except Exception as e:
    print(f"Error al conectar a la base de datos: {e}")

# Configuración del commit en bloques
commit_chunk = 10000
update_counter = 0

# Recorrer cada fila del CSV
for index, row in df.iterrows():
    try:
        ruota = row['Ruota']
        numeros = [int(row['1°']), int(row['2°']), int(row['3°']), int(row['4°']), int(row['5°'])]

        # Generar todas las combinaciones de "ambos" (pares ordenados) y ejecutar el update
        for i in range(len(numeros) - 1):
            for j in range(i + 1, len(numeros)):
                n1, n2 = numeros[i], numeros[j]

                update_query = sql.SQL("""
                    UPDATE ambos_matrix
                    SET ocurrencias = ocurrencias + 1
                    WHERE ruota = %s AND numero_1 = %s AND numero_2 = %s
                """)

                cursor.execute(update_query, (ruota, n1, n2))
                update_counter += 1

        # Commit en bloques
        if update_counter >= commit_chunk:
            conn.commit()
            print(f"Commit realizado después de {update_counter} actualizaciones.")
            update_counter = 0

    except Exception as e:
        print(f"Error procesando la fila {index+1}: {e}")

# Commit final para las actualizaciones restantes
if update_counter > 0:
    conn.commit()
    print(f"Commit final realizado con {update_counter} actualizaciones restantes.")

# Cerrar conexión
cursor.close()
conn.close()
print("Proceso completado. La columna 'ocurrencias' en la tabla 'ambos_matrix' ha sido actualizada correctamente.")
```
